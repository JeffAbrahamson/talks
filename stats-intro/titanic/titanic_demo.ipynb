```python
import pandas as pd
import numpy as np
import seaborn as sns
import pprint
pd.options.display.width = 180
import matplotlib.pyplot as plt

df = pd.read_table('train.csv', sep=",")
```

```python
df.head()
```

```python
%matplotlib inline  
sns.barplot('Survived', data=df)
```

```python
sns.barplot('Sex', data=df)
```

```python
sns.factorplot(x="Sex", y="Survived", data=df)
```

```python
df['sex_'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df.head()
```

```python
from sklearn import tree
import numpy as np
feature_names = np.array(['sex_']) 
df_train_Y = df[['Survived']]
train_Y = df_train_Y.values.ravel()  # turn from 2D to 1D

df_train_X = df[feature_names]
train_X = df_train_X.values

clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_X,train_Y)
output = clf.predict(train_X)
```

```python
output.shape
print(output[:10])
```

```python
clf.score(train_X, train_Y)
```

Generating the plot in the next cell requires that graphviz (also known simply as "dot") be installed.  You can skip this step if you don't have dot and don't want to install it.  Of course, you'll miss out on the pretty visualisation.

```python
import io

with open("titanic.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names=feature_names)
import os
os.system("dot -Tpng titanic.dot -o titanic.png")

from IPython.display import Image
Image(filename='titanic.png')
```

```python
# make 10-fold cross val
import sklearn
kf=sklearn.cross_validation.KFold(n=len(train_X), n_folds=10)
# print the scores for each of the 10 folds against the test set
cv_scores=sklearn.cross_validation.cross_val_score(clf, train_X, y=train_Y, cv=kf)
#plt.plot(cv_scores, 'o')
#plt.ylim((0,1))
#plt.title("Cross validation scores, {} +/- {}".format(cv_scores.mean(), cv_scores.std()))
print("Mean cross validation (held-out data) score:", np.mean(cv_scores))
```

# Decision Tree max_depth=5 (hope to prevent overfitting)

```python
feature_names = np.array(['sex_', 'Fare'])  #['Pclass', 'sex_', 'Fare', 'Parch', 'SibSp', 'emb_s_', 'emb_c_', 'emb_q_'])

df_train_X = df[feature_names]

train_X = df_train_X.values
df_train_X = df[feature_names]

clf = tree.DecisionTreeClassifier(max_depth=5)
clf = clf.fit(train_X,train_Y)
output2 = clf.predict(train_X)
#print((output==output2).all())
clf.score(train_X, train_Y)
```

```python
import io

with open("titanic.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names=feature_names)
import os
os.system("dot -Tpng titanic.dot -o titanic.png")

from IPython.display import Image
Image(filename='titanic.png')
```
